```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    accuracy_score, precision_recall_fscore_support, confusion_matrix, classification_report
)
from ultralytics import YOLO
import cv2
import random
import os

# ==============================
# 1. Load model & dataset
# ==============================
model = YOLO("runs/detect/wordbased-asl + asl-dataset-p9yw8 v2/weights/best.pt")  # change path if needed
data_yaml = "wordbased-asl + asl-dataset-p9yw8 v2/data 2.yaml"  # your dataset yaml

# Evaluate on validation set
metrics = model.val(data=data_yaml, split="val")  # YOLO's built-in metrics
print(metrics)

# ==============================
# 2. Custom Evaluation on Validation Images
# ==============================
from ultralytics.data.utils import check_det_dataset
dataset_info = check_det_dataset(data_yaml)
val_path = dataset_info["val"]

# Collect predictions and ground truth
all_labels, all_preds = [], []
image_dir = os.path.join(val_path, "images")
label_dir = os.path.join(val_path, "labels")

for label_file in os.listdir(label_dir):
    if not label_file.endswith(".txt"):
        continue
    img_file = label_file.replace(".txt", ".jpg")
    img_path = os.path.join(image_dir, img_file)
    label_path = os.path.join(label_dir, label_file)

    # Ground truth labels
    gt_classes = []
    with open(label_path, "r") as f:
        for line in f:
            class_id = int(line.split()[0])
            gt_classes.append(class_id)

    # Run prediction
    results = model(img_path, conf=0.5)
    pred_classes = [int(b.cls.cpu().numpy()) for b in results[0].boxes]

    # Pad or truncate predictions to match ground truths
    if len(pred_classes) < len(gt_classes):
        pred_classes.extend([-1] * (len(gt_classes) - len(pred_classes)))
    elif len(pred_classes) > len(gt_classes):
        pred_classes = pred_classes[:len(gt_classes)]

    all_labels.extend(gt_classes)
    all_preds.extend(pred_classes)

# ==============================
# 3. Metrics
# ==============================
target_names = dataset_info["names"]

print("\nClassification Report:")
print(classification_report(all_labels, all_preds, target_names=target_names, zero_division=0))

# Overall accuracy
acc = accuracy_score(all_labels, all_preds)
print(f"\nOverall Accuracy: {acc:.4f}")

# Per-class precision, recall, F1
prec, rec, f1, _ = precision_recall_fscore_support(all_labels, all_preds, zero_division=0)
for i, cls in enumerate(target_names):
    print(f"Class {cls} -> Precision: {prec[i]:.2f}, Recall: {rec[i]:.2f}, F1: {f1[i]:.2f}")

# ==============================
# 4. Confusion Matrix
# ==============================
cm = confusion_matrix(all_labels, all_preds, labels=list(range(len(target_names))))
plt.figure(figsize=(14,12))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names, yticklabels=target_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix - Validation Set")
plt.show()

# ==============================
# 5. Training Curves (Loss/Accuracy)
# ==============================
# Load YOLO's training results file
results_file = "runs/detect/wordbased-asl + asl-dataset-p9yw8 v2/results.csv"
if os.path.exists(results_file):
    import pandas as pd
    df = pd.read_csv(results_file)

    plt.figure(figsize=(10,6))
    plt.plot(df["epoch"], df["train/box_loss"], label="Train Box Loss")
    plt.plot(df["epoch"], df["train/cls_loss"], label="Train Class Loss")
    plt.plot(df["epoch"], df["metrics/mAP50"], label="mAP@0.5 (Val)")
    plt.xlabel("Epoch")
    plt.ylabel("Loss / mAP")
    plt.legend()
    plt.title("Training Dynamics")
    plt.show()

# ==============================
# 6. Visual Inspection of Predictions
# ==============================
sample_images = random.sample(os.listdir(image_dir), 5)
for img_file in sample_images:
    img_path = os.path.join(image_dir, img_file)
    results = model(img_path, conf=0.5)
    annotated = results[0].plot()
    cv2.imshow("Prediction Example", annotated)
    cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import os
import cv2
import yaml
import glob
import time
import numpy as np
from collections import deque
from ultralytics import YOLO
import mediapipe as mp

# ============================================================================
# CONFIGURATION
# ============================================================================
class Config:
    DATA_YAML_PATH = r"C:\Users\kusal\OneDrive\Documents\Uni stuff\Year 5 Sem 2\FYP-B\Datasets\asl-dataset-p9yw8 v2\data.yaml"
    MODEL_PATH = "runs/detect/asl-dataset-p9yw8 v2/weights/best.pt"
    CONF_THRESHOLD = 0.6  # Increased for better accuracy
    PAUSE_DURATION = 1.5  # Seconds of no detection = word confirmed
    MAX_SENTENCE_LENGTH = 10
    DISPLAY_SCALE = 1.5
    
    # ASL Grammar patterns (simplified)
    ASL_PATTERNS = {
        'statement': ['SUBJECT', 'OBJECT', 'VERB'],
        'question': ['QUESTION_WORD', 'SUBJECT', 'OBJECT'],
    }

# ============================================================================
# HAND KEYPOINT ANALYZER (Error Detection)
# ============================================================================
class HandKeypointAnalyzer:
    def __init__(self):
        self.mp_hands = mp.solutions.hands.Hands(
            static_image_mode=False,
            max_num_hands=2,
            min_detection_confidence=0.9,
            min_tracking_confidence=0.9
        )
        self.mp_drawing = mp.solutions.drawing_utils
        self.reference_keypoints = {}  # Will store reference data
        
    def extract_keypoints(self, frame):
        """Extract hand landmarks from frame"""
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = self.mp_hands.process(frame_rgb)
        
        if results.multi_hand_landmarks:
            keypoints = []
            for hand_landmarks in results.multi_hand_landmarks:
                kp = np.array([[lm.x, lm.y, lm.z] for lm in hand_landmarks.landmark])
                keypoints.append(kp)
            return keypoints, results.multi_hand_landmarks
        return None, None
    
    def save_reference_keypoints(self, sign_name, keypoints):
        """Save reference keypoints for a sign"""
        self.reference_keypoints[sign_name] = keypoints
    
    def calculate_finger_angles(self, keypoints):
        """Calculate angles of fingers for hand shape analysis"""
        if keypoints is None or len(keypoints) == 0:
            return None
        
        kp = keypoints[0]  # Use first hand
        angles = {}
        
        # Finger tip indices: thumb=4, index=8, middle=12, ring=16, pinky=20
        # Base indices: thumb=2, index=5, middle=9, ring=13, pinky=17
        fingers = {
            'thumb': (2, 3, 4),
            'index': (5, 6, 7, 8),
            'middle': (9, 10, 11, 12),
            'ring': (13, 14, 15, 16),
            'pinky': (17, 18, 19, 20)
        }
        
        for finger_name, indices in fingers.items():
            if len(indices) >= 3:
                # Calculate angle at middle joint
                p1, p2, p3 = kp[indices[0]], kp[indices[1]], kp[indices[2]]
                v1 = p1 - p2
                v2 = p3 - p2
                angle = np.arccos(np.clip(np.dot(v1, v2) / 
                                         (np.linalg.norm(v1) * np.linalg.norm(v2)), -1, 1))
                angles[finger_name] = np.degrees(angle)
        
        return angles
    
    def compare_hand_shape(self, user_keypoints, reference_keypoints):
        """Compare user's hand shape with reference"""
        if user_keypoints is None or reference_keypoints is None:
            return None
        
        user_angles = self.calculate_finger_angles(user_keypoints)
        ref_angles = self.calculate_finger_angles(reference_keypoints)
        
        if user_angles is None or ref_angles is None:
            return None
        
        errors = []
        total_diff = 0
        
        for finger in ['thumb', 'index', 'middle', 'ring', 'pinky']:
            if finger in user_angles and finger in ref_angles:
                diff = abs(user_angles[finger] - ref_angles[finger])
                total_diff += diff
                
                if diff > 30:  # 30 degree threshold
                    if user_angles[finger] > ref_angles[finger]:
                        errors.append(f"{finger.capitalize()} too bent")
                    else:
                        errors.append(f"{finger.capitalize()} too extended")
        
        # Calculate similarity score (0-100)
        similarity = max(0, 100 - (total_diff / 5))  # Normalize by 5 fingers
        
        return {
            'similarity': similarity,
            'errors': errors
        }
    
    def analyze_hand_orientation(self, keypoints):
        """Analyze palm orientation"""
        if keypoints is None or len(keypoints) == 0:
            return None
        
        kp = keypoints[0]
        # Use wrist (0), middle finger base (9), and pinky base (17)
        wrist = kp[0]
        middle_base = kp[9]
        pinky_base = kp[17]
        
        # Calculate palm normal vector
        v1 = middle_base - wrist
        v2 = pinky_base - wrist
        normal = np.cross(v1[:2], v2[:2])  # 2D cross product
        
        # Determine orientation
        angle = np.arctan2(normal, v1[0])
        return np.degrees(angle)

# ============================================================================
# SENTENCE BUILDER
# ============================================================================
class SentenceBuilder:
    def __init__(self, class_names):
        self.class_names = class_names
        self.sign_buffer = []
        self.current_word = None
        self.last_detection_time = None
        self.word_start_time = None
        self.detection_counts = {}
        self.min_detections = 5  # Minimum detections to confirm a word
        
    def add_detection(self, class_id, confidence):
        """Add a detection and build sentence"""
        current_time = time.time()
        
        # If no detection for PAUSE_DURATION, confirm word
        if (self.last_detection_time and 
            current_time - self.last_detection_time > Config.PAUSE_DURATION):
            self._confirm_current_word()
            self.detection_counts = {}
        
        # Count detections
        if class_id not in self.detection_counts:
            self.detection_counts[class_id] = 0
        self.detection_counts[class_id] += 1
        
        # Update current word if enough detections
        if self.detection_counts[class_id] >= self.min_detections:
            self.current_word = (class_id, confidence)
        
        self.last_detection_time = current_time
        
    def _confirm_current_word(self):
        """Confirm and add current word to sentence"""
        if self.current_word and len(self.sign_buffer) < Config.MAX_SENTENCE_LENGTH:
            class_id, confidence = self.current_word
            word = self.class_names[class_id]
            
            # Avoid duplicates
            if not self.sign_buffer or self.sign_buffer[-1] != word:
                self.sign_buffer.append(word)
            
            self.current_word = None
    
    def get_current_word(self):
        """Get the word being signed right now"""
        if self.current_word:
            class_id, _ = self.current_word
            return self.class_names[class_id]
        return None
    
    def get_sentence(self):
        """Get current sentence"""
        return " ".join(self.sign_buffer)
    
    def clear_last_word(self):
        """Remove last word from sentence"""
        if self.sign_buffer:
            self.sign_buffer.pop()
    
    def clear_sentence(self):
        """Clear entire sentence"""
        self.sign_buffer = []
        self.current_word = None
        self.detection_counts = {}
    
    def validate_grammar(self):
        """Basic ASL grammar validation"""
        if len(self.sign_buffer) < 2:
            return True, ""
        
        # Simple validation: check for common patterns
        # This is a placeholder - you'd expand based on your signs
        sentence = " ".join(self.sign_buffer)
        
        # Example: Check if question words are at the start
        question_words = ['WHO', 'WHAT', 'WHERE', 'WHEN', 'WHY', 'HOW']
        if any(word in self.sign_buffer for word in question_words):
            if self.sign_buffer[0] not in question_words:
                return False, "ASL questions typically start with question word"
        
        return True, ""

# ============================================================================
# MAIN APPLICATION
# ============================================================================
class ASLTeachingApp:
    def __init__(self):
        # Load configuration
        with open(Config.DATA_YAML_PATH, "r") as f:
            self.data_yaml = yaml.safe_load(f)
        self.class_names = self.data_yaml['names']
        
        # Initialize components
        self.model = YOLO(Config.MODEL_PATH)
        self.keypoint_analyzer = HandKeypointAnalyzer()
        self.sentence_builder = SentenceBuilder(self.class_names)
        
        # Load reference images
        self.reference_list = self._load_reference_images()
        self.current_reference_idx = 0
        
        # Mode: 'teaching' or 'sentence'
        self.mode = 'teaching'
        
        # For reference keypoints capture
        self.capturing_reference = False
        
    def _load_reference_images(self):
        """Load one reference image per class"""
        image_per_class = {}
        for split in ['train', 'valid', 'test']:
            if split not in self.data_yaml:
                continue
            images_dir = self.data_yaml[split]
            label_dir = images_dir.replace('images', 'labels')
            image_files = glob.glob(os.path.join(images_dir, "*.jpg")) + \
                         glob.glob(os.path.join(images_dir, "*.png"))
            
            for img_path in image_files:
                label_path = os.path.join(label_dir, 
                                         os.path.splitext(os.path.basename(img_path))[0] + ".txt")
                if os.path.exists(label_path):
                    with open(label_path, "r") as lf:
                        for line in lf:
                            class_id = int(line.split()[0])
                            if class_id not in image_per_class:
                                image_per_class[class_id] = img_path
                            if len(image_per_class) == len(self.class_names):
                                break
            
            if len(image_per_class) == len(self.class_names):
                break
        
        return [(image_per_class[c], self.class_names[c]) 
                for c in sorted(image_per_class.keys())]
    
    def draw_ui(self, frame, annotated_frame, detection_info):
        """Draw UI elements on frame"""
        # Ensure the info panel width matches the annotated_frame width.
        # annotated_frame may include a concatenated reference image, so use its width.
        h, w = annotated_frame.shape[:2]
        
        # Create info panel
        panel_height = 200
        panel = np.zeros((panel_height, w, 3), dtype=np.uint8)
        panel[:] = (40, 40, 40)
        
        y_offset = 30
        
        # Mode indicator
        mode_text = f"Mode: {'TEACHING' if self.mode == 'teaching' else 'SENTENCE BUILDING'}"
        cv2.putText(panel, mode_text, (10, y_offset), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
        y_offset += 35
        
        if self.mode == 'sentence':
            # Current word being signed
            current_word = self.sentence_builder.get_current_word()
            if current_word:
                cv2.putText(panel, f"Signing: {current_word}", (10, y_offset),
                           cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
            y_offset += 30
            
            # Current sentence
            sentence = self.sentence_builder.get_sentence()
            cv2.putText(panel, f"Sentence: {sentence}", (10, y_offset),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
            y_offset += 30
            
            # Grammar validation
            is_valid, grammar_msg = self.sentence_builder.validate_grammar()
            if not is_valid:
                cv2.putText(panel, f"Grammar: {grammar_msg}", (10, y_offset),
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
        
        else:  # teaching mode
            if detection_info:
                similarity = detection_info.get('similarity', 0)
                errors = detection_info.get('errors', [])
                
                # Similarity score
                color = (0, 255, 0) if similarity > 80 else (0, 255, 255) if similarity > 60 else (0, 0, 255)
                cv2.putText(panel, f"Accuracy: {similarity:.1f}%", (10, y_offset),
                           cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)
                y_offset += 35
                
                # Error feedback
                if errors:
                    cv2.putText(panel, "Corrections needed:", (10, y_offset),
                               cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 0), 1)
                    y_offset += 25
                    for i, error in enumerate(errors[:3]):  # Show max 3 errors
                        cv2.putText(panel, f"- {error}", (20, y_offset),
                                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)
                        y_offset += 20
        
        # Controls
        y_offset = panel_height - 35
        cv2.putText(panel, "Controls: [T]each/Sentence | [SPACE]Next Sign | [C]lear | [Q]uit", 
                   (10, y_offset), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (200, 200, 200), 1)
        
        # Combine with frame
        combined = np.vstack([annotated_frame, panel])
        return combined
    
    def run(self):
        """Main application loop"""
        cap = cv2.VideoCapture(0)
        
        print("=" * 60)
        print("ASL Teaching System - Controls:")
        print("  [T] - Toggle Teaching/Sentence mode")
        print("  [SPACE] - Next reference sign (teaching mode)")
        print("  [C] - Clear sentence (sentence mode)")
        print("  [R] - Capture reference keypoints (teaching mode)")
        print("  [Q] - Quit")
        print("=" * 60)
        
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            
            # Run YOLO inference
            results = self.model(frame, imgsz=416, conf=Config.CONF_THRESHOLD, verbose=False)
            annotated_frame = results[0].plot()
            
            # Extract hand keypoints
            user_keypoints, hand_landmarks = self.keypoint_analyzer.extract_keypoints(frame)
            
            # Process detections
            detection_info = None
            if len(results[0].boxes) > 0:
                # Get highest confidence detection
                boxes = results[0].boxes
                best_idx = boxes.conf.argmax()
                class_id = int(boxes.cls[best_idx])
                confidence = float(boxes.conf[best_idx])
                
                if self.mode == 'sentence':
                    self.sentence_builder.add_detection(class_id, confidence)
                else:  # teaching mode
                    current_sign = self.reference_list[self.current_reference_idx][1]
                    if self.class_names[class_id] == current_sign:
                        # Compare with reference keypoints if available
                        if current_sign in self.keypoint_analyzer.reference_keypoints:
                            ref_kp = self.keypoint_analyzer.reference_keypoints[current_sign]
                            detection_info = self.keypoint_analyzer.compare_hand_shape(
                                user_keypoints, ref_kp)
            
            # Display reference image in teaching mode
            if self.mode == 'teaching':
                ref_path, ref_label = self.reference_list[self.current_reference_idx]
                ref_img = cv2.imread(ref_path)
                h = annotated_frame.shape[0]
                ref_aspect = ref_img.shape[1] / ref_img.shape[0]
                ref_img = cv2.resize(ref_img, (int(h * ref_aspect), h))
                
                # Add label
                cv2.putText(ref_img, f"Sign: {ref_label}", (10, 40),
                           cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 0), 3)
                
                annotated_frame = cv2.hconcat([ref_img, annotated_frame])
            
            # Draw UI
            display_frame = self.draw_ui(frame, annotated_frame, detection_info)
            
            # Scale and show
            if Config.DISPLAY_SCALE != 1.0:
                display_frame = cv2.resize(display_frame, (0, 0), 
                                          fx=Config.DISPLAY_SCALE, 
                                          fy=Config.DISPLAY_SCALE)
            
            cv2.imshow("ASL Teaching System", display_frame)
            
            # Handle keyboard input
            key = cv2.waitKey(1) & 0xFF
            if key == ord('q'):
                break
            elif key == ord('t'):
                self.mode = 'sentence' if self.mode == 'teaching' else 'teaching'
                print(f"Switched to {self.mode.upper()} mode")
            elif key == ord(' ') and self.mode == 'teaching':
                self.current_reference_idx = (self.current_reference_idx + 1) % len(self.reference_list)
            elif key == ord('c') and self.mode == 'sentence':
                self.sentence_builder.clear_sentence()
                print("Sentence cleared")
            elif key == ord('r') and self.mode == 'teaching':
                # Capture reference keypoints
                if user_keypoints:
                    current_sign = self.reference_list[self.current_reference_idx][1]
                    self.keypoint_analyzer.save_reference_keypoints(current_sign, user_keypoints)
                    print(f"Captured reference keypoints for {current_sign}")
        
        cap.release()
        cv2.destroyAllWindows()

# ============================================================================
# RUN APPLICATION
# ============================================================================
if __name__ == "__main__":
    app = ASLTeachingApp()
    app.run()
```

```python
# asl_teaching_feedback.py
"""
ASL teaching app with per-finger comparison between reference image keypoints
and live webcam keypoints. Displays side-by-side reference + live frames,
draws MediaPipe landmarks, and gives per-finger "bend/extend" feedback.
"""
import os
import cv2
import yaml
import glob
import time
import numpy as np
from collections import deque
from ultralytics import YOLO
import mediapipe as mp
from pathlib import Path

# =============================================================================
# Config
# =============================================================================
class Config:
    DATA_YAML_PATH = r"C:\Users\kusal\OneDrive\Documents\Uni stuff\Year 5 Sem 2\FYP-B\Datasets\asl-dataset-p9yw8 v2\data.yaml"
    MODEL_PATH = "runs/detect/asl-dataset-p9yw8 v2/weights/best.pt"
    CONF_THRESHOLD = 0.7
    PAUSE_DURATION = 1.5
    MAX_SENTENCE_LENGTH = 10
    DISPLAY_SCALE = 1.2
    SEQ_LEN = 64
    REFERENCE_FILE = "reference_keypoints.npz"
    EMA_ALPHA = 0.6
    MATCH_ANGLE_THRESHOLD = 20.0   # degrees: below => OK, above => feedback
    # visual params
    REF_WIDTH = 360
    LIVE_WIDTH = 640
    PANEL_HEIGHT = 300

# =============================================================================
# Keypoint utility / MediaPipe Holistic wrapper
# =============================================================================
class HandKeypointAnalyzer:
    def __init__(self):
        self.holistic = mp.solutions.holistic.Holistic(static_image_mode=False, model_complexity=0)
        self.mp_drawing = mp.solutions.drawing_utils
        self.mp_styles = mp.solutions.drawing_styles
        self.reference_keypoints = {}          # {sign_name: {'left':arr,'right':arr,'pose':arr}}
        self.ema_alpha = Config.EMA_ALPHA
        self.last_kp = None
        self._load_references_on_disk_or_images()

    def _load_references_on_disk_or_images(self):
        # attempt to load precomputed references
        p = Path(Config.REFERENCE_FILE)
        if p.exists():
            try:
                data = np.load(p, allow_pickle=True)
                for k in data.files:
                    self.reference_keypoints[k] = data[k].tolist()
                print(f"Loaded {len(self.reference_keypoints)} reference(s) from {p}")
            except Exception as e:
                print("Could not load reference keypoints:", e)
        # note: actual mapping from class name -> image path handled at ASLTeachingApp init,
        # we will compute missing references on demand via extract_keypoints_from_image()

    def save_references_to_disk(self):
        try:
            np.savez_compressed(Config.REFERENCE_FILE, **self.reference_keypoints)
            print("Saved reference_keypoints to", Config.REFERENCE_FILE)
        except Exception as e:
            print("Failed saving references:", e)

    def extract_keypoints_from_image(self, image_path):
        """
        Run holistic on a static image and return kp_dict {'left','right','pose'} (arrays or None).
        Coordinates are normalized (x,y,z) as MediaPipe returns.
        """
        img = cv2.imread(str(image_path))
        if img is None:
            return None
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        # for static image, better to use static_image_mode True temporarily
        with mp.solutions.holistic.Holistic(static_image_mode=True, model_complexity=0) as h:
            res = h.process(img_rgb)
        out = {'left': None, 'right': None, 'pose': None}
        if getattr(res, 'pose_landmarks', None):
            out['pose'] = np.array([[lm.x, lm.y, lm.z] for lm in res.pose_landmarks.landmark], dtype=np.float32)
        if getattr(res, 'left_hand_landmarks', None):
            out['left'] = np.array([[lm.x, lm.y, lm.z] for lm in res.left_hand_landmarks.landmark], dtype=np.float32)
        if getattr(res, 'right_hand_landmarks', None):
            out['right'] = np.array([[lm.x, lm.y, lm.z] for lm in res.right_hand_landmarks.landmark], dtype=np.float32)
        return out

    def extract_keypoints(self, frame):
        """Process live frame (non-static) and smooth via EMA."""
        img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        res = self.holistic.process(img_rgb)
        out = {'left': None, 'right': None, 'pose': None}
        if getattr(res, 'pose_landmarks', None):
            out['pose'] = np.array([[lm.x, lm.y, lm.z] for lm in res.pose_landmarks.landmark], dtype=np.float32)
        if getattr(res, 'left_hand_landmarks', None):
            out['left'] = np.array([[lm.x, lm.y, lm.z] for lm in res.left_hand_landmarks.landmark], dtype=np.float32)
        if getattr(res, 'right_hand_landmarks', None):
            out['right'] = np.array([[lm.x, lm.y, lm.z] for lm in res.right_hand_landmarks.landmark], dtype=np.float32)
        out = self._smooth_keypoints(out)
        return out, res

    def _smooth_keypoints(self, kp_dict):
        if self.last_kp is None:
            self.last_kp = {}
            for k in ['left','right','pose']:
                v = kp_dict.get(k)
                self.last_kp[k] = v.copy() if v is not None else None
            return kp_dict
        sm = {}
        a = self.ema_alpha
        for k in ['left','right','pose']:
            cur = kp_dict.get(k)
            last = self.last_kp.get(k)
            if cur is None:
                sm[k] = last
            else:
                if last is None:
                    sm[k] = cur
                else:
                    try:
                        sm[k] = a*cur + (1.0-a)*last
                    except Exception:
                        sm[k] = cur
            self.last_kp[k] = sm[k]
        return sm

    def draw_landmarks(self, image, res, draw_face=False):
        """Draw pose and hand landmarks on the BGR image (in place)."""
        try:
            if getattr(res, 'pose_landmarks', None):
                self.mp_drawing.draw_landmarks(
                    image, res.pose_landmarks, mp.solutions.holistic.POSE_CONNECTIONS,
                    landmark_drawing_spec=self.mp_styles.get_default_pose_landmarks_style()
                )
            if getattr(res, 'left_hand_landmarks', None):
                self.mp_drawing.draw_landmarks(
                    image, res.left_hand_landmarks, mp.solutions.holistic.HAND_CONNECTIONS,
                    landmark_drawing_spec=self.mp_styles.get_default_hand_landmarks_style()
                )
            if getattr(res, 'right_hand_landmarks', None):
                self.mp_drawing.draw_landmarks(
                    image, res.right_hand_landmarks, mp.solutions.holistic.HAND_CONNECTIONS,
                    landmark_drawing_spec=self.mp_styles.get_default_hand_landmarks_style()
                )
            if draw_face and getattr(res, 'face_landmarks', None):
                self.mp_drawing.draw_landmarks(image, res.face_landmarks, mp.solutions.holistic.FACEMESH_TESSELATION,
                                               landmark_drawing_spec=self.mp_styles.get_default_face_mesh_landmarks_style())
        except Exception as e:
            # don't crash display
            print("Landmark draw error:", e)
        return image

    # finger-angle utilities
    def calculate_finger_angles(self, hand_kp):
        """Return dict finger_name -> angle at middle joint (deg). If hand_kp is None -> None."""
        if hand_kp is None:
            return None
        kp = hand_kp
        angles = {}
        mapping = {
            'thumb': (2,3,4),
            'index': (5,6,7),
            'middle': (9,10,11),
            'ring': (13,14,15),
            'pinky': (17,18,19)
        }
        for name,(a,b,c) in mapping.items():
            p1 = kp[a][:2]; p2 = kp[b][:2]; p3 = kp[c][:2]
            v1 = p1 - p2; v2 = p3 - p2
            denom = (np.linalg.norm(v1)*np.linalg.norm(v2))
            if denom < 1e-6:
                angles[name] = 0.0
            else:
                cosv = np.clip(np.dot(v1,v2)/denom, -1.0, 1.0)
                angles[name] = float(np.degrees(np.arccos(cosv)))
        return angles

    def compare_hand_shape_verbose(self, live_kp_dict, ref_kp_dict, angle_threshold=Config.MATCH_ANGLE_THRESHOLD):
        """
        Compare live vs reference per finger. Returns dict:
          {'left'/'right': {'similarity':float, 'per_finger': {finger: {'ref_angle', 'live_angle','diff','advice'}} }, 'best_side': 'left'/'right' }
        If a side missing, it will be omitted.
        Advice = 'ok' or 'extend' or 'bend'
        """
        if live_kp_dict is None or ref_kp_dict is None:
            return None

        results = {}
        sides = ['left','right']
        for side in sides:
            live = live_kp_dict.get(side)
            ref = ref_kp_dict.get(side)
            if live is None or ref is None:
                continue
            live_angles = self.calculate_finger_angles(live)
            ref_angles = self.calculate_finger_angles(ref)
            if live_angles is None or ref_angles is None:
                continue
            per_finger = {}
            total_diff = 0.0; comps = 0
            for f in ['thumb','index','middle','ring','pinky']:
                la = live_angles.get(f, 0.0)
                ra = ref_angles.get(f, 0.0)
                diff = la - ra  # positive -> live more bent than ref (assuming angle increases with bend)
                absdiff = abs(diff)
                comps += 1
                total_diff += absdiff
                if absdiff <= angle_threshold:
                    advice = 'ok'
                else:
                    # if live angle larger (more bent) -> suggest extend; else suggest bend
                    advice = 'bend' if diff > 0 else 'extend'
                per_finger[f] = {'ref_angle': ra, 'live_angle': la, 'diff': diff, 'absdiff': absdiff, 'advice': advice}
            avg_diff = total_diff / max(1, comps)
            similarity = max(0.0, 100.0 * (1.0 - (avg_diff / 90.0)))
            results[side] = {'similarity': similarity, 'avg_absdiff': avg_diff, 'per_finger': per_finger}
        # choose best side by similarity if both present
        best_side = None
        if results:
            best_side = max(results.items(), key=lambda x: x[1]['similarity'])[0]
        return {'sides': results, 'best_side': best_side}

# =============================================================================
# Sentence builder (unchanged)
# =============================================================================
class SentenceBuilder:
    def __init__(self, class_names):
        self.class_names = class_names
        self.sign_buffer = []
        self.current_word = None
        self.last_detection_time = None
        self.detection_counts = {}
        self.min_detections = 3
        self.confirm_pause = Config.PAUSE_DURATION
    def add_detection(self, class_id, confidence):
        t = time.time()
        if self.last_detection_time and (t - self.last_detection_time > self.confirm_pause):
            self._confirm_current_word()
            self.detection_counts = {}
        self.detection_counts[class_id] = self.detection_counts.get(class_id,0) + 1
        if self.detection_counts[class_id] >= self.min_detections:
            self.current_word = (class_id, confidence)
        self.last_detection_time = t
    def _confirm_current_word(self):
        if self.current_word and len(self.sign_buffer) < Config.MAX_SENTENCE_LENGTH:
            class_id, _ = self.current_word
            word = self.class_names[class_id]
            if not self.sign_buffer or self.sign_buffer[-1] != word:
                self.sign_buffer.append(word)
            self.current_word = None
    def get_current_word(self):
        if self.current_word:
            class_id, _ = self.current_word
            return self.class_names[class_id]
        return None
    def get_sentence(self):
        return " ".join(self.sign_buffer)
    def clear_sentence(self):
        self.sign_buffer = []
        self.current_word = None
        self.detection_counts = {}
        self.last_detection_time = None
    def validate_grammar(self):
        if len(self.sign_buffer) < 2:
            return True, ""
        question_words = ['WHO','WHAT','WHERE','WHEN','WHY','HOW']
        if any(w in self.sign_buffer for w in question_words):
            if self.sign_buffer[0] not in question_words:
                return False, "ASL questions typically start with question word"
        return True, ""

# =============================================================================
# Main app
# =============================================================================
class ASLTeachingApp:
    def __init__(self):
        with open(Config.DATA_YAML_PATH, 'r') as f:
            self.data_yaml = yaml.safe_load(f)
        self.class_names = self.data_yaml['names']
        self.model = YOLO(Config.MODEL_PATH)
        self.keypoint_analyzer = HandKeypointAnalyzer()
        self.sentence_builder = SentenceBuilder(self.class_names)
        # reference list maps class_id -> image_path (one image per class)
        self.reference_list = self._load_reference_images_one_per_class()
        # ensure references' keypoints exist (compute if missing)
        self._ensure_reference_keypoints()
        self.current_reference_idx = 0
        self.mode = 'teaching'
        self.seq_buffer = deque(maxlen=Config.SEQ_LEN)

    def _load_reference_images_one_per_class(self):
        img_per_class = {}
        # check common splits 'valid','test','train'
        for split in ['valid','test','train']:
            if split not in self.data_yaml:
                continue
            images_dir = self.data_yaml[split]
            if not os.path.isabs(images_dir):
                images_dir = os.path.join(os.path.dirname(os.path.abspath(Config.DATA_YAML_PATH)), images_dir)
            label_dir = images_dir.replace('images','labels')
            files = glob.glob(os.path.join(images_dir, '*.jpg')) + glob.glob(os.path.join(images_dir, '*.png')) + glob.glob(os.path.join(images_dir, '*.jpeg'))
            for p in files:
                lab = os.path.join(label_dir, os.path.splitext(os.path.basename(p))[0] + '.txt')
                if os.path.exists(lab):
                    with open(lab,'r') as lf:
                        line = lf.readline().strip()
                        if not line: continue
                        cid = int(line.split()[0])
                        if cid not in img_per_class:
                            img_per_class[cid] = p
                            # stop early if all classes found
                            if len(img_per_class) == len(self.class_names):
                                break
            if len(img_per_class) == len(self.class_names):
                break
        # produce list of (image_path, class_name) sorted by class index
        result = []
        for c in sorted(img_per_class.keys()):
            result.append((img_per_class[c], self.class_names[c]))
        return result

    def _ensure_reference_keypoints(self):
        # For each reference image, compute and store keypoints under the class name if not already stored
        for img_path, class_name in self.reference_list:
            if class_name in self.keypoint_analyzer.reference_keypoints:
                continue
            kp = self.keypoint_analyzer.extract_keypoints_from_image(img_path)
            if kp is not None:
                # store mapping by class_name
                self.keypoint_analyzer.reference_keypoints[class_name] = kp
                print(f"Computed reference keypoints for {class_name} from {img_path}")
        # persist
        self.keypoint_analyzer.save_references_to_disk()

    def _resize_keep_aspect(self, img, width=None, height=None):
        if img is None:
            return None
        h,w = img.shape[:2]
        if width is not None:
            scale = width / float(w)
            return cv2.resize(img, (width, int(h*scale)))
        if height is not None:
            scale = height / float(h)
            return cv2.resize(img, (int(w*scale), height))
        return img

    def _draw_finger_feedback_on_frame(self, frame_bgr, comparison_result, side='right'):
        """
        Draw small colored circles near the 'middle joint' and text advice for each finger.
        Assumes comparison_result is the output of compare_hand_shape_verbose and 'best_side' exists.
        We'll compute screen coords from MediaPipe normalized coordinates in live_kp_dict.
        """
        if comparison_result is None:
            return frame_bgr
        sides = comparison_result.get('sides', {})
        if side not in sides:
            return frame_bgr
        per = sides[side]['per_finger']
        # But we need the live hand keypoints to draw positions; pass them by including them in results? 
        # To keep things simple, this function expects the live normalized keypoints to be used externally to get screen coords.
        # Instead, we'll rely on that the caller will overlay circles using known 2D coords provided below.
        return frame_bgr

    def draw_ui_panel(self, annotated_live, ref_img, comp_verbose):
        # Build the right-side panel (dark) containing textual feedback
        w = annotated_live.shape[1]
        panel_h = Config.PANEL_HEIGHT
        panel = np.zeros((panel_h, w, 3), dtype=np.uint8) + 30
        y = 25
        title = f"Mode: {'TEACHING' if self.mode=='teaching' else 'SENTENCE'}"
        cv2.putText(panel, title, (10,y), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (200,200,255), 2)
        y += 30
        # show reference sign name
        if len(self.reference_list) > 0:
            _, sign_name = self.reference_list[self.current_reference_idx]
            cv2.putText(panel, f"Reference: {sign_name}", (10,y), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (180,255,180), 2)
        y += 28
        # show similarity & per-finger advice
        if comp_verbose is None:
            cv2.putText(panel, "No comparison available", (10,y), cv2.FONT_HERSHEY_SIMPLEX, 0.55, (200,200,200), 1)
        else:
            best = comp_verbose.get('best_side', None)
            sides = comp_verbose.get('sides', {})
            if best and best in sides:
                sim = sides[best]['similarity']
                cv2.putText(panel, f"Best side: {best}  Similarity: {sim:.1f}%", (10,y), cv2.FONT_HERSHEY_SIMPLEX, 0.55, (200,255,200), 1)
                y += 22
                per = sides[best]['per_finger']
                for finger in ['thumb','index','middle','ring','pinky']:
                    info = per.get(finger)
                    if not info: continue
                    adv = info['advice']
                    col = (50,200,50) if adv=='ok' else (50,50,200) if adv=='extend' else (0,0,255)
                    txt = f"{finger.capitalize():6s}: {adv} (Δ={info['absdiff']:.1f}°)"
                    cv2.putText(panel, txt, (10,y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, col, 1)
                    y += 20
            else:
                cv2.putText(panel, "No hand(s) in reference/live", (10,y), cv2.FONT_HERSHEY_SIMPLEX, 0.55, (200,200,200), 1)
        # controls
        cv2.putText(panel, "Controls: [T] Toggle Mode  [SPACE] Next Ref  [R] Save Ref  [Q] Quit", (10, panel_h-20),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.45, (200,200,200), 1)
        return panel

    def _norm_to_screen(self, norm_xy, image_w, image_h):
        # norm_xy = (x,y) in [0,1]. If None, return (-1,-1)
        if norm_xy is None:
            return -1, -1
        x = int(norm_xy[0] * image_w)
        y = int(norm_xy[1] * image_h)
        return x, y

    def run(self):
        cap = cv2.VideoCapture(0)
        print("Starting ASL teaching app. Press Q to quit.")
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            orig_h, orig_w = frame.shape[:2]
            # YOLO inference
            results = self.model(frame, imgsz=416, conf=Config.CONF_THRESHOLD, verbose=False)
            # Use model's plotted frame or original depending on alignment; prefer original for landmark overlay consistency
            live_frame = frame.copy()  # we'll overlay landmarks and boxes manually
            # extract mediapipe keypoints & results
            live_kp_dict, mp_res = self.keypoint_analyzer.extract_keypoints(frame)
            # draw MP landmarks on live_frame for visualization
            live_vis = live_frame.copy()
            live_vis = self.keypoint_analyzer.draw_landmarks(live_vis, mp_res, draw_face=False)
            # draw YOLO boxes (simplified)
            try:
                boxes = results[0].boxes
                for b in boxes:
                    # b.xyxy, b.conf, b.cls
                    xyxy = b.xyxy.cpu().numpy().astype(int)[0]
                    conf = float(b.conf.cpu().numpy()[0]) if hasattr(b.conf, 'cpu') else float(b.conf)
                    cls = int(b.cls.cpu().numpy()[0]) if hasattr(b.cls, 'cpu') else int(b.cls)
                    label = f"{self.class_names[cls]} {conf:.2f}"
                    cv2.rectangle(live_vis, (xyxy[0], xyxy[1]), (xyxy[2], xyxy[3]), (0,200,0), 2)
                    cv2.putText(live_vis, label, (xyxy[0], max(0,xyxy[1]-6)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,200,0), 2)
            except Exception:
                pass

            # Reference image and keypoints
            if len(self.reference_list) > 0:
                ref_img_path, ref_name = self.reference_list[self.current_reference_idx]
                ref_kp = self.keypoint_analyzer.reference_keypoints.get(ref_name, None)
                ref_img = cv2.imread(ref_img_path)
                if ref_img is None:
                    ref_img = np.zeros((Config.REF_WIDTH, Config.REF_WIDTH, 3), dtype=np.uint8)
                ref_img = self._resize_keep_aspect(ref_img, width=Config.REF_WIDTH)
                # Draw landmarks on reference image using static extraction (recompute a small static res)
                # For the ref HUD we recompute a static result so we can use mp drawing utils
                ref_rgb = cv2.cvtColor(ref_img, cv2.COLOR_BGR2RGB)
                # compute result for drawing
                with mp.solutions.holistic.Holistic(static_image_mode=True, model_complexity=0) as h:
                    ref_res = h.process(ref_rgb)
                ref_vis = ref_img.copy()
                # ref_vis = self.keypoint_analyzer.draw_landmarks(ref_vis, ref_res, draw_face=False)
            else:
                ref_vis = np.zeros((Config.REF_WIDTH, Config.REF_WIDTH, 3), dtype=np.uint8)
                ref_name = "N/A"
                ref_kp = None

            # If in teaching mode and class matches reference, compare shapes and produce feedback
            comp_verbose = None
            try:
                # choose best detected class (highest conf) if any
                boxes = results[0].boxes
                if len(boxes) > 0:
                    confs = boxes.conf.cpu().numpy()
                    cls_arr = boxes.cls.cpu().numpy()
                    idx = int(confs.argmax())
                    best_cls = int(cls_arr[idx])
                    best_conf = float(confs[idx])
                    # if reference exists and label matches
                    if len(self.reference_list) > 0 and self.class_names[best_cls] == ref_name:
                        comp_verbose = self.keypoint_analyzer.compare_hand_shape_verbose(live_kp_dict, ref_kp)
            except Exception:
                comp_verbose = None

            # Build composite: left=ref_vis, center=live_vis resized, right=panel
            live_resized = self._resize_keep_aspect(live_vis, width=Config.LIVE_WIDTH)
            panel = self.draw_ui_panel(live_resized, ref_vis, comp_verbose)
            # overlay per-joint colored markers onto live_resized using live_kp_dict and comp_verbose
            if comp_verbose and comp_verbose.get('best_side') and live_kp_dict:
                best_side = comp_verbose.get('best_side')
                per = comp_verbose['sides'][best_side]['per_finger']
                # pick the hand to draw from live_kp_dict
                hand_kp = live_kp_dict.get(best_side)
                if hand_kp is not None:
                    h_h, h_w = live_resized.shape[:2]
                    # draw small circles for each finger middle joint (index b)
                    map_mid = {'thumb':3, 'index':6, 'middle':10, 'ring':14, 'pinky':18}
                    for finger, mid_idx in map_mid.items():
                        norm_xy = (hand_kp[mid_idx][0], hand_kp[mid_idx][1])  # normalized
                        sx, sy = self._norm_to_screen(norm_xy, h_w, h_h)
                        info = per.get(finger)
                        if info:
                            color = (0,200,0) if info['advice']=='ok' else (0,255,255) if info['advice']=='extend' else (0,0,255)
                            cv2.circle(live_resized, (sx,sy), 10, color, -1)
                            # small text
                            adv = info['advice']
                            cv2.putText(live_resized, adv[0].upper(), (sx+10, sy-6), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 1)

            # Create final horizontal arrangement
            # Left: ref_vis, Middle: live_resized, Right: panel
            # ensure heights match
            target_h = max(ref_vis.shape[0], live_resized.shape[0], panel.shape[0])
            def pad_to_h(img, h):
                ih, iw = img.shape[:2]
                if ih == h:
                    return img
                pad = np.zeros((h-ih, iw, 3), dtype=np.uint8) + 10
                return np.vstack([img, pad])
            ref_p = pad_to_h(ref_vis, target_h)
            live_p = pad_to_h(live_resized, target_h)
            panel_p = pad_to_h(panel, target_h)
            composed = np.hstack([ref_p, live_p, panel_p])

            # optional scaling for display
            if Config.DISPLAY_SCALE != 1.0:
                composed = cv2.resize(composed, (0,0), fx=Config.DISPLAY_SCALE, fy=Config.DISPLAY_SCALE)

            cv2.imshow("ASL Teaching (ref | live | feedback)", composed)
            key = cv2.waitKey(1) & 0xFF
            if key == ord('q'):
                break
            elif key == ord('t'):
                self.mode = 'sentence' if self.mode == 'teaching' else 'teaching'
            elif key == ord(' ') and len(self.reference_list) > 0:
                self.current_reference_idx = (self.current_reference_idx + 1) % len(self.reference_list)
            elif key == ord('r') and len(self.reference_list) > 0:
                # capture the current live keypoints as new reference for current sign
                _, sign_name = self.reference_list[self.current_reference_idx]
                self.keypoint_analyzer.reference_keypoints[sign_name] = live_kp_dict
                self.keypoint_analyzer.save_references_to_disk()
                print(f"Saved live keypoints as reference for {sign_name}")

        cap.release()
        cv2.destroyAllWindows()

# =============================================================================
# Run
# =============================================================================
if __name__ == "__main__":
    app = ASLTeachingApp()
    app.run()
```

```python
# asl_teaching_feedback_presentation_full.py
"""
ASL teaching app with:
 - green markers for correct fingers, red for incorrect
 - publication PNG exporter (300 DPI) using green/red scheme
 - session summary CSV of mean per-finger errors
 - headless / record-only CLI mode with automatic snapshots
"""
import os
import cv2
import yaml
import glob
import time
import json
import numpy as np
import argparse
import csv
from collections import deque
from ultralytics import YOLO
import mediapipe as mp
from pathlib import Path
import matplotlib.pyplot as plt

# -------------------------
# Config
# -------------------------
class Config:
    DATA_YAML_PATH = r"C:\Users\kusal\OneDrive\Documents\Uni stuff\Year 5 Sem 2\FYP-B\Datasets\asl-dataset-p9yw8 v2\data.yaml"   # <- update to your data.yaml path
    MODEL_PATH = "runs/detect/asl-dataset-p9yw8 v2/weights/best.pt"
    CONF_THRESHOLD = 0.7
    PAUSE_DURATION = 1.5
    MAX_SENTENCE_LENGTH = 10
    DISPLAY_SCALE = 1.0
    SEQ_LEN = 64
    REFERENCE_FILE = "reference_keypoints.npz"
    EMA_ALPHA = 0.6
    MATCH_ANGLE_THRESHOLD = 20.0   # degrees
    REF_WIDTH = 360
    LIVE_WIDTH = 640
    PANEL_HEIGHT = 300
    SAVE_DIR = "demo_artifacts"
    VIDEO_FPS = 30
os.makedirs(Config.SAVE_DIR, exist_ok=True)

# -------------------------
# Keypoint analyzer (MediaPipe Holistic wrapper)
# -------------------------
class HandKeypointAnalyzer:
    def __init__(self):
        self.holistic = mp.solutions.holistic.Holistic(static_image_mode=False, model_complexity=0)
        self.mp_drawing = mp.solutions.drawing_utils
        self.mp_styles = mp.solutions.drawing_styles
        self.reference_keypoints = {}
        self.ema_alpha = Config.EMA_ALPHA
        self.last_kp = None
        self._load_references()

    def _load_references(self):
        p = Path(Config.REFERENCE_FILE)
        if p.exists():
            try:
                data = np.load(p, allow_pickle=True)
                for k in data.files:
                    self.reference_keypoints[k] = data[k].tolist()
                print(f"Loaded {len(self.reference_keypoints)} reference(s) from {p}")
            except Exception as e:
                print("Could not load reference keypoints:", e)

    def save_references_to_disk(self):
        try:
            np.savez_compressed(Config.REFERENCE_FILE, **self.reference_keypoints)
            print("Saved reference_keypoints to", Config.REFERENCE_FILE)
        except Exception as e:
            print("Failed saving references:", e)

    def extract_keypoints_from_image(self, image_path):
        img = cv2.imread(str(image_path))
        if img is None: return None
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        with mp.solutions.holistic.Holistic(static_image_mode=True, model_complexity=0) as h:
            res = h.process(img_rgb)
        out = {'left': None, 'right': None, 'pose': None}
        if getattr(res, 'pose_landmarks', None):
            out['pose'] = np.array([[lm.x,lm.y,lm.z] for lm in res.pose_landmarks.landmark], dtype=np.float32)
        if getattr(res, 'left_hand_landmarks', None):
            out['left'] = np.array([[lm.x,lm.y,lm.z] for lm in res.left_hand_landmarks.landmark], dtype=np.float32)
        if getattr(res, 'right_hand_landmarks', None):
            out['right'] = np.array([[lm.x,lm.y,lm.z] for lm in res.right_hand_landmarks.landmark], dtype=np.float32)
        return out

    def extract_keypoints(self, frame):
        rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        res = self.holistic.process(rgb)
        out = {'left': None, 'right': None, 'pose': None}
        if getattr(res, 'pose_landmarks', None):
            out['pose'] = np.array([[lm.x,lm.y,lm.z] for lm in res.pose_landmarks.landmark], dtype=np.float32)
        if getattr(res, 'left_hand_landmarks', None):
            out['left'] = np.array([[lm.x,lm.y,lm.z] for lm in res.left_hand_landmarks.landmark], dtype=np.float32)
        if getattr(res, 'right_hand_landmarks', None):
            out['right'] = np.array([[lm.x,lm.y,lm.z] for lm in res.right_hand_landmarks.landmark], dtype=np.float32)
        out = self._smooth_keypoints(out)
        return out, res

    def _smooth_keypoints(self, kp_dict):
        if self.last_kp is None:
            self.last_kp = {}
            for k in ['left','right','pose']:
                v = kp_dict.get(k)
                self.last_kp[k] = v.copy() if v is not None else None
            return kp_dict
        sm = {}
        a = self.ema_alpha
        for k in ['left','right','pose']:
            cur = kp_dict.get(k); last = self.last_kp.get(k)
            if cur is None:
                sm[k] = last
            else:
                if last is None:
                    sm[k] = cur
                else:
                    try:
                        sm[k] = a*cur + (1.0-a)*last
                    except Exception:
                        sm[k] = cur
            self.last_kp[k] = sm[k]
        return sm

    def draw_landmarks(self, image, res, draw_face=False):
        try:
            if getattr(res, 'pose_landmarks', None):
                self.mp_drawing.draw_landmarks(image, res.pose_landmarks, mp.solutions.holistic.POSE_CONNECTIONS,
                                               landmark_drawing_spec=self.mp_styles.get_default_pose_landmarks_style())
            if getattr(res, 'left_hand_landmarks', None):
                self.mp_drawing.draw_landmarks(image, res.left_hand_landmarks, mp.solutions.holistic.HAND_CONNECTIONS,
                                               landmark_drawing_spec=self.mp_styles.get_default_hand_landmarks_style())
            if getattr(res, 'right_hand_landmarks', None):
                self.mp_drawing.draw_landmarks(image, res.right_hand_landmarks, mp.solutions.holistic.HAND_CONNECTIONS,
                                               landmark_drawing_spec=self.mp_styles.get_default_hand_landmarks_style())
            if draw_face and getattr(res, 'face_landmarks', None):
                self.mp_drawing.draw_landmarks(image, res.face_landmarks, mp.solutions.holistic.FACEMESH_TESSELATION,
                                               landmark_drawing_spec=self.mp_styles.get_default_face_mesh_landmarks_style())
        except Exception as e:
            print("Landmark draw error:", e)
        return image

    def calculate_finger_angles(self, hand_kp):
        if hand_kp is None: return None
        kp = hand_kp
        angles = {}
        mapping = {'thumb':(2,3,4),'index':(5,6,7),'middle':(9,10,11),'ring':(13,14,15),'pinky':(17,18,19)}
        for name,(a,b,c) in mapping.items():
            p1 = kp[a][:2]; p2 = kp[b][:2]; p3 = kp[c][:2]
            v1 = p1 - p2; v2 = p3 - p2
            denom = (np.linalg.norm(v1)*np.linalg.norm(v2))
            if denom < 1e-6:
                angles[name] = 0.0
            else:
                cosv = np.clip(np.dot(v1,v2)/denom, -1.0, 1.0)
                angles[name] = float(np.degrees(np.arccos(cosv)))
        return angles

    def compare_hand_shape_verbose(self, live_kp_dict, ref_kp_dict, angle_threshold=Config.MATCH_ANGLE_THRESHOLD):
        if live_kp_dict is None or ref_kp_dict is None: return None
        results = {}
        for side in ['left','right']:
            live = live_kp_dict.get(side); ref = ref_kp_dict.get(side)
            if live is None or ref is None: continue
            live_angles = self.calculate_finger_angles(live); ref_angles = self.calculate_finger_angles(ref)
            if live_angles is None or ref_angles is None: continue
            per_finger = {}; total_diff = 0.0; comps = 0
            for f in ['thumb','index','middle','ring','pinky']:
                la = live_angles.get(f,0.0); ra = ref_angles.get(f,0.0)
                diff = la - ra; absdiff = abs(diff)
                comps += 1; total_diff += absdiff
                if absdiff <= angle_threshold: advice = 'ok'
                else: advice = 'extend' if diff > 0 else 'bend'
                per_finger[f] = {'ref_angle':ra, 'live_angle':la, 'diff':diff, 'absdiff':absdiff, 'advice':advice}
            avg_diff = total_diff / max(1, comps)
            similarity = max(0.0, 100.0*(1.0 - (avg_diff/90.0)))
            results[side] = {'similarity':similarity, 'avg_absdiff':avg_diff, 'per_finger':per_finger}
        best_side = None
        if results:
            best_side = max(results.items(), key=lambda x:x[1]['similarity'])[0]
        return {'sides':results, 'best_side':best_side}

# -------------------------
# Utility: publication PNG exporter (green/red)
# -------------------------
def export_publication_png(image_path, meta_json_path, out_path, dpi=300):
    try:
        img = cv2.imread(image_path)
        if img is None:
            raise RuntimeError("image not found")
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        with open(meta_json_path, 'r') as jf:
            meta = json.load(jf)
    except Exception as e:
        print("Export failed:", e); return False

    per = meta.get('per_finger', {})
    finger_order = ['thumb','index','middle','ring','pinky']
    absvals = [per.get(f, {}).get('absdiff', 0.0) for f in finger_order]
    advs = [per.get(f, {}).get('advice', 'ok') for f in finger_order]

    fig = plt.figure(figsize=(10,5), dpi=dpi)
    gs = fig.add_gridspec(1, 2, width_ratios=[2,1], wspace=0.12)
    ax_img = fig.add_subplot(gs[0,0]); ax_bar = fig.add_subplot(gs[0,1])

    ax_img.imshow(img); ax_img.axis('off')
    title = f"Ref: {meta.get('ref_name','N/A')}  Detected: {meta.get('best_cls', 'N/A')}  Conf: {meta.get('best_conf', 0.0):.2f}"
    ax_img.set_title(title, fontsize=10)

    # horizontal bar chart: green = ok, red = incorrect (extend/bend)
    colors = []
    for a in advs:
        if a == 'ok': colors.append('#2ca02c')      # green
        else: colors.append('#d62728')              # red
    y = np.arange(len(finger_order))
    ax_bar.barh(y, absvals, color=colors)
    ax_bar.set_yticks(y); ax_bar.set_yticklabels([f.capitalize() for f in finger_order])
    ax_bar.set_xlabel('Absolute angle error (°)')
    ax_bar.set_xlim(0, 90)
    sim = meta.get('similarity', None)
    if sim is not None:
        ax_bar.set_title(f"Similarity: {sim:.1f}%")
    fig.suptitle("ASL per-finger comparison — publication export", fontsize=12)
    plt.tight_layout()
    try:
        fig.savefig(out_path, dpi=dpi)
        plt.close(fig)
        print("Saved publication PNG:", out_path)
        return True
    except Exception as e:
        print("Failed saving publication PNG:", e)
        return False

# -------------------------
# Main app (colors: green=ok, red=incorrect)
# -------------------------
class ASLTeachingApp:
    def __init__(self, data_yaml_path=Config.DATA_YAML_PATH, model_path=Config.MODEL_PATH):
        with open(data_yaml_path, 'r') as f:
            self.data_yaml = yaml.safe_load(f)
        self.class_names = self.data_yaml['names']
        self.model = YOLO(model_path)
        self.kpa = HandKeypointAnalyzer()
        self.sbuilder = SentenceBuilder(self.class_names) if 'SentenceBuilder' in globals() else None
        self.reference_list = self._load_reference_images_one_per_class()
        self._ensure_reference_keypoints()
        self.current_reference_idx = 0
        self.mode = 'teaching'
        self.seq_buffer = deque(maxlen=Config.SEQ_LEN)
        self.recording = False
        self.video_writer = None
        self.last_infer_time = None
        self.session_snaps = []

    def _load_reference_images_one_per_class(self):
        img_per_class = {}
        for split in ['valid','test','train']:
            if split not in self.data_yaml:
                continue
            images_dir = self.data_yaml[split]
            if not os.path.isabs(images_dir):
                images_dir = os.path.join(os.path.dirname(os.path.abspath(Config.DATA_YAML_PATH)), images_dir)
            label_dir = images_dir.replace('images','labels')
            files = glob.glob(os.path.join(images_dir, '*.*'))
            for p in files:
                if not p.lower().endswith(('.jpg','.png','.jpeg')): continue
                lab = os.path.join(label_dir, os.path.splitext(os.path.basename(p))[0] + '.txt')
                if os.path.exists(lab):
                    with open(lab,'r') as lf:
                        line = lf.readline().strip()
                        if not line: continue
                        cid = int(line.split()[0])
                        if cid not in img_per_class:
                            img_per_class[cid] = p
                            if len(img_per_class) == len(self.class_names):
                                break
            if len(img_per_class) == len(self.class_names):
                break
        result = []
        for c in sorted(img_per_class.keys()):
            result.append((img_per_class[c], self.class_names[c]))
        return result

    def _ensure_reference_keypoints(self):
        for img_path, class_name in self.reference_list:
            if class_name in self.kpa.reference_keypoints: continue
            kp = self.kpa.extract_keypoints_from_image(img_path)
            if kp is not None:
                self.kpa.reference_keypoints[class_name] = kp
                print(f"Computed reference keypoints for {class_name} from {img_path}")
        self.kpa.save_references_to_disk()

    def _resize_keep_aspect(self, img, width=None, height=None):
        if img is None: return None
        h,w = img.shape[:2]
        if width is not None:
            scale = width / float(w)
            return cv2.resize(img, (width, int(h*scale)))
        if height is not None:
            scale = height / float(h)
            return cv2.resize(img, (int(w*scale), height))
        return img

    def _norm_to_screen(self, norm_xy, image_w, image_h):
        if norm_xy is None: return -1,-1
        x = int(np.clip(norm_xy[0], 0.0, 1.0) * image_w)
        y = int(np.clip(norm_xy[1], 0.0, 1.0) * image_h)
        return x, y

    def _save_snapshot(self, composed, meta):
        t = time.strftime("%Y%m%d_%H%M%S")
        img_path = os.path.join(Config.SAVE_DIR, f"snapshot_{t}.png")
        json_path = os.path.join(Config.SAVE_DIR, f"snapshot_{t}.json")
        cv2.imwrite(img_path, composed)
        with open(json_path, 'w') as jf:
            json.dump(meta, jf, indent=2)
        self.session_snaps.append((img_path, json_path, meta))
        print(f"Saved snapshot {img_path}")
        return img_path, json_path

    def _finalize_recording_session(self, out_prefix=None, save_pub_pngs=False):
        if not self.session_snaps:
            print("No snapshots in session.")
            return
        finger_order = ['thumb','index','middle','ring','pinky']
        accum = {f:[] for f in finger_order}
        for (img_p, json_p, meta) in self.session_snaps:
            per = meta.get('per_finger', {})
            for f in finger_order:
                val = per.get(f, {}).get('absdiff', None)
                if val is None:
                    continue
                accum[f].append(float(val))
        means = {f: (np.mean(accum[f]) if len(accum[f])>0 else None) for f in finger_order}
        t = time.strftime("%Y%m%d_%H%M%S")
        out_name = out_prefix if out_prefix else f"session_summary_{t}"
        csv_path = os.path.join(Config.SAVE_DIR, f"{out_name}.csv")
        with open(csv_path, 'w', newline='') as cf:
            writer = csv.writer(cf)
            writer.writerow(['finger','mean_absdiff_deg','samples'])
            for f in finger_order:
                arr = accum[f]
                writer.writerow([f, f"{(np.mean(arr)):.3f}" if len(arr)>0 else "", len(arr)])
        print("Saved session summary CSV:", csv_path)
        if save_pub_pngs:
            for img_p, json_p, meta in self.session_snaps:
                pub_out = img_p.replace("snapshot_", "publication_")
                pub_out = os.path.splitext(pub_out)[0] + ".png"
                export_publication_png(img_p, json_p, pub_out, dpi=300)
        self.session_snaps = []

    def run(self, headless=False, record_only_duration=0, snapshot_interval=1.0, save_pub_pngs=False, out_dir=None):
        cap = cv2.VideoCapture(0)
        if not cap.isOpened():
            raise RuntimeError("Could not open webcam")
        last_time = time.time()
        last_snapshot = time.time()
        start_time = time.time()
        auto_end_time = start_time + record_only_duration if record_only_duration > 0 else None
        if record_only_duration > 0:
            self.recording = True
            print(f"Auto-recording ON for {record_only_duration}s; snapshots every {snapshot_interval}s")

        while True:
            ret, frame = cap.read()
            if not ret:
                break
            t0 = time.time()
            results = self.model(frame, imgsz=416, conf=Config.CONF_THRESHOLD, verbose=False)
            t1 = time.time(); infer_dt = t1 - t0; self.last_infer_time = infer_dt

            live_kp_dict, mp_res = self.kpa.extract_keypoints(frame)
            live_vis = frame.copy()
            live_vis = self.kpa.draw_landmarks(live_vis, mp_res, draw_face=False)

            best_conf = 0.0; best_cls = None
            try:
                boxes = results[0].boxes
                if len(boxes) > 0:
                    confs = boxes.conf.cpu().numpy(); cls_arr = boxes.cls.cpu().numpy()
                    idx = int(confs.argmax()); best_conf = float(confs[idx]); best_cls = int(cls_arr[idx])
                    for b in boxes:
                        xyxy = b.xyxy.cpu().numpy().astype(int)[0]
                        conf = float(b.conf.cpu().numpy()[0]) if hasattr(b.conf, 'cpu') else float(b.conf)
                        cls = int(b.cls.cpu().numpy()[0]) if hasattr(b.cls, 'cpu') else int(b.cls)
                        label = f"{self.class_names[cls]} {conf:.2f}"
                        cv2.rectangle(live_vis, (xyxy[0], xyxy[1]), (xyxy[2], xyxy[3]), (0,200,0), 2)
                        cv2.putText(live_vis, label, (max(0,xyxy[0]), max(0,xyxy[1]-6)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,200,0), 2)
            except Exception:
                pass

            if len(self.reference_list) > 0:
                ref_img_path, ref_name = self.reference_list[self.current_reference_idx]
                ref_kp = self.kpa.reference_keypoints.get(ref_name, None)
                ref_img = cv2.imread(ref_img_path)
                if ref_img is None:
                    ref_img = np.zeros((Config.REF_WIDTH, Config.REF_WIDTH,3), dtype=np.uint8)
                ref_img = self._resize_keep_aspect(ref_img, width=Config.REF_WIDTH)
                ref_rgb = cv2.cvtColor(ref_img, cv2.COLOR_BGR2RGB)
                with mp.solutions.holistic.Holistic(static_image_mode=True, model_complexity=0) as h:
                    ref_res = h.process(ref_rgb)
                ref_vis = ref_img.copy()
                ref_vis = self.kpa.draw_landmarks(ref_vis, ref_res, draw_face=False)
            else:
                ref_vis = np.zeros((Config.REF_WIDTH, Config.REF_WIDTH, 3), dtype=np.uint8)
                ref_name = "N/A"; ref_kp = None

            comp_verbose = None
            if best_cls is not None and len(self.reference_list) > 0 and self.class_names[best_cls] == ref_name:
                comp_verbose = self.kpa.compare_hand_shape_verbose(live_kp_dict, ref_kp)

            live_resized = self._resize_keep_aspect(live_vis, width=Config.LIVE_WIDTH)
            panel = np.zeros((Config.PANEL_HEIGHT, live_resized.shape[1], 3), dtype=np.uint8) + 30
            fps = 1.0/(time.time() - last_time + 1e-8); last_time = time.time()
            sim_txt = "N/A"
            per_finger = {}
            if comp_verbose and comp_verbose.get('best_side'):
                best_side = comp_verbose.get('best_side')
                sim = comp_verbose['sides'][best_side]['similarity']; sim_txt = f"{sim:.1f}%"
                per_finger = comp_verbose['sides'][best_side]['per_finger']
            cv2.putText(panel, f"Similarity: {sim_txt}", (10,30), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (220,220,220),2)
            cv2.putText(panel, f"Model conf: {best_conf:.2f}", (10,58), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (200,200,200),1)
            cv2.putText(panel, f"FPS: {fps:.1f}", (10,86), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (200,200,200),1)
            if self.last_infer_time:
                cv2.putText(panel, f"Infer(ms): {self.last_infer_time*1000:.1f}", (10,112), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (200,200,200),1)

            # legend (green/red)
            x = panel.shape[1] - 180; y = 10
            cv2.rectangle(panel, (x,y), (x+170, y+60), (20,20,20), -1)
            cv2.putText(panel, "Legend:", (x+8, y+18), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (230,230,230), 1)
            cv2.rectangle(panel, (x+8, y+28), (x+28, y+48), (0,200,0), -1); cv2.putText(panel, " OK", (x+34, y+44), cv2.FONT_HERSHEY_SIMPLEX, 0.45, (230,230,230),1)
            cv2.rectangle(panel, (x+80, y+28), (x+100, y+48), (0,0,255), -1); cv2.putText(panel, " Incorrect", (x+106, y+44), cv2.FONT_HERSHEY_SIMPLEX, 0.45, (230,230,230),1)

            # per-finger bar area (green/red)
            chart_x = 10; chart_y = panel.shape[0] - 110; chart_w = panel.shape[1] - 20; chart_h = 90
            cv2.rectangle(panel, (chart_x, chart_y), (chart_x + chart_w, chart_y + chart_h), (10,10,10), -1)
            finger_order = ['thumb','index','middle','ring','pinky']
            bar_h = int(chart_h / len(finger_order))
            for i,f in enumerate(finger_order):
                val = per_finger.get(f, {}).get('absdiff', 0.0) if per_finger else 0.0
                norm = float(np.clip(val / 90.0, 0.0, 1.0))
                x0 = chart_x + 8; y0 = chart_y + i * bar_h + 8
                x1 = x0 + int((chart_w-40) * norm); y1 = y0 + bar_h - 12
                cv2.rectangle(panel, (x0,y0), (chart_x+chart_w-24, y1), (40,40,40), -1)
                # green if ok, red otherwise (extend/bend both red)
                color = (0,200,0) if per_finger and per_finger.get(f, {}).get('advice') == 'ok' else (0,0,255)
                cv2.rectangle(panel, (x0,y0), (x1,y1), color, -1)
                cv2.putText(panel, f"{f[:3].upper()} {val:.0f}°", (chart_x+chart_w-20, y0 + (bar_h-12)//2 + 6), cv2.FONT_HERSHEY_SIMPLEX, 0.45, (230,230,230),1, cv2.LINE_AA)

            # overlay per-joint markers on live_resized (green/red)
            if comp_verbose and comp_verbose.get('best_side') and live_kp_dict:
                best_side = comp_verbose.get('best_side')
                per = comp_verbose['sides'][best_side]['per_finger']
                hand_kp = live_kp_dict.get(best_side)
                if hand_kp is not None:
                    h_h, h_w = live_resized.shape[:2]
                    map_mid = {'thumb':3, 'index':6, 'middle':10, 'ring':14, 'pinky':18}
                    for finger, mid_idx in map_mid.items():
                        try:
                            nx = float(hand_kp[mid_idx][0]); ny = float(hand_kp[mid_idx][1])
                        except Exception:
                            continue
                        sx, sy = self._norm_to_screen((nx,ny), h_w, h_h)
                        info = per.get(finger)
                        if info:
                            # green for correct, red for incorrect (both extend/bend)
                            if info['advice'] == 'ok':
                                marker_color = (0,200,0)
                            else:
                                marker_color = (0,0,255)
                            cv2.circle(live_resized, (sx,sy), 14, (0,0,0), thickness=-1)
                            cv2.circle(live_resized, (sx,sy), 12, marker_color, thickness=-1)
                            letter = 'O' if info['advice']=='ok' else 'X'
                            cv2.putText(live_resized, letter, (sx-8, sy+6), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255,255,255), 2, cv2.LINE_AA)
                            dtxt = f"{info['diff']:+.1f}°"
                            cv2.putText(live_resized, dtxt, (sx+18, sy-12), cv2.FONT_HERSHEY_SIMPLEX, 0.45, (230,230,230), 1, cv2.LINE_AA)

            # assemble composite
            target_h = max(ref_vis.shape[0], live_resized.shape[0], panel.shape[0])
            def pad_to_h(img,h):
                ih, iw = img.shape[:2]
                if ih==h: return img
                pad = np.zeros((h-ih, iw,3), dtype=np.uint8) + 10
                return np.vstack([img, pad])
            ref_p = pad_to_h(ref_vis, target_h); live_p = pad_to_h(live_resized, target_h); panel_p = pad_to_h(panel, target_h)
            composed = np.hstack([ref_p, live_p, panel_p])

            # auto-snapshot if recording and snapshot_interval elapsed
            now = time.time()
            if self.recording and snapshot_interval>0 and (now - last_snapshot) >= snapshot_interval:
                meta = {
                    'timestamp': time.strftime("%Y-%m-%d %H:%M:%S"),
                    'ref_name': ref_name,
                    'best_cls': self.class_names[best_cls] if best_cls is not None else None,
                    'best_conf': float(best_conf),
                    'similarity': comp_verbose['sides'][comp_verbose['best_side']]['similarity'] if comp_verbose and comp_verbose.get('best_side') else None,
                    'per_finger': per_finger
                }
                img_p, json_p = self._save_snapshot(composed, meta)
                if save_pub_pngs:
                    pub_out = os.path.splitext(img_p)[0] + "_publication.png"
                    export_publication_png(img_p, json_p, pub_out, dpi=300)
                last_snapshot = now

            if not headless:
                show = composed.copy()
                if Config.DISPLAY_SCALE != 1.0:
                    show = cv2.resize(show, (0,0), fx=Config.DISPLAY_SCALE, fy=Config.DISPLAY_SCALE)
                cv2.imshow("ASL Teaching (ref | live | feedback)", show)
                key = cv2.waitKey(1) & 0xFF
            else:
                key = 0xFF

            if key == ord('q'):
                if self.recording:
                    self.recording = False
                    self._finalize_recording_session(save_pub_pngs=save_pub_pngs)
                break
            elif key == ord(' '):
                if len(self.reference_list)>0: self.current_reference_idx = (self.current_reference_idx + 1) % len(self.reference_list)
            elif key == ord('r'):
                if len(self.reference_list)>0:
                    _, sign_name = self.reference_list[self.current_reference_idx]
                    self.kpa.reference_keypoints[sign_name] = live_kp_dict
                    self.kpa.save_references_to_disk()
                    print(f"Saved live keypoints as reference for {sign_name}")
            elif key == ord('s'):
                meta = {
                    'timestamp': time.strftime("%Y-%m-%d %H:%M:%S"),
                    'ref_name': ref_name,
                    'best_cls': self.class_names[best_cls] if best_cls is not None else None,
                    'best_conf': float(best_conf),
                    'similarity': comp_verbose['sides'][comp_verbose['best_side']]['similarity'] if comp_verbose and comp_verbose.get('best_side') else None,
                    'per_finger': per_finger
                }
                img_p, json_p = self._save_snapshot(composed, meta)
                if save_pub_pngs:
                    pub_out = os.path.splitext(img_p)[0] + "_publication.png"
                    export_publication_png(img_p, json_p, pub_out, dpi=300)
            elif key == ord('v'):
                self.recording = not self.recording
                if not self.recording:
                    self._finalize_recording_session(save_pub_pngs=save_pub_pngs)
                    print("Recording stopped and session finalized.")
                else:
                    print("Recording started (manual toggle).")
            if auto_end_time is not None and time.time() >= auto_end_time:
                print("Auto record duration reached; finalizing.")
                if self.recording:
                    self.recording = False
                    self._finalize_recording_session(save_pub_pngs=save_pub_pngs)
                break

        cap.release()
        if not headless:
            cv2.destroyAllWindows()

# -------------------------
# Minimal SentenceBuilder stub for compatibility
# -------------------------
class SentenceBuilder:
    def __init__(self, class_names):
        self.class_names = class_names
        self.sign_buffer = []
        self.current_word = None
        self.last_detection_time = None
        self.detection_counts = {}
        self.min_detections = 3
        self.confirm_pause = Config.PAUSE_DURATION
    def add_detection(self, class_id, confidence):
        t = time.time()
        if self.last_detection_time and (t - self.last_detection_time > self.confirm_pause):
            self._confirm_current_word()
            self.detection_counts = {}
        self.detection_counts[class_id] = self.detection_counts.get(class_id,0)+1
        if self.detection_counts[class_id] >= self.min_detections:
            self.current_word = (class_id, confidence)
        self.last_detection_time = t
    def _confirm_current_word(self):
        if self.current_word and len(self.sign_buffer) < Config.MAX_SENTENCE_LENGTH:
            class_id,_ = self.current_word
            word = self.class_names[class_id]
            if not self.sign_buffer or self.sign_buffer[-1] != word:
                self.sign_buffer.append(word)
            self.current_word = None

# -------------------------
# CLI runner
# -------------------------
def parse_args_and_run():
    parser = argparse.ArgumentParser()
    parser.add_argument('--data', default=Config.DATA_YAML_PATH, help='Path to data.yaml')
    parser.add_argument('--model', default=Config.MODEL_PATH, help='YOLO model path')
    parser.add_argument('--headless', action='store_true', help='No GUI; run headless (useful for servers)')
    parser.add_argument('--record-only', type=int, default=0, help='Record-only mode for N seconds (auto snapshots)')
    parser.add_argument('--snapshot-interval', type=float, default=1.0, help='Seconds between auto-snapshots when recording')
    parser.add_argument('--save-pub-pngs', action='store_true', help='Export publication PNGs for saved snapshots (300 DPI)')
    
    # Filter out Jupyter kernel arguments
    import sys
    args_to_parse = [arg for arg in sys.argv[1:] if not arg.startswith('--f=')]
    args = parser.parse_args(args_to_parse)

    app = ASLTeachingApp(data_yaml_path=args.data, model_path=args.model)
    print("Starting app with args:", args)
    app.run(headless=args.headless, record_only_duration=args.record_only,
            snapshot_interval=args.snapshot_interval, save_pub_pngs=args.save_pub_pngs)

if __name__ == "__main__":
    parse_args_and_run()
```
